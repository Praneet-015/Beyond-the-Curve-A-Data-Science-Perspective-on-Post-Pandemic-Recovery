# Beyond the Curve – A Data Science Perspective on Post-Pandemic Recovery

### Load and Parse Date Column

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error
import math
from sklearn.preprocessing import MinMaxScaler
from prophet import Prophet
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense


def load_and_prepare_data(filepath):
    """
    Load the dataset and convert the 'date' column to datetime format.
    """
    df = pd.read_csv(filepath)
    df['date'] = pd.to_datetime(df['date'])
    df.sort_values(by='date', inplace=True)
    return df
```

### Drop Columns with All Missing Values

```python
def drop_fully_missing_columns(df):
    """
    Remove columns that are completely empty (all values are NaN).
    """
    missing_all = df.columns[df.isnull().sum() == len(df)]
    return df.drop(columns=missing_all)
```

### Fill Missing Values in Stringency Index

```python
def fill_stringency_index(df):
    """
    Replace missing 'stringency_index' values with the column median.
    """
    if 'stringency_index' in df.columns:
        median_val = df['stringency_index'].median()
        df['stringency_index'].fillna(median_val, inplace=True)
    return df
```

### Set Pre-Vaccine Values to Zero

```python
def reset_pre_vaccine_data(df):
    """
    For dates before 2021-01-16, set vaccine-related columns to 0.
    """
    vaccine_cols = [
        'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated',
        'total_boosters', 'new_vaccinations', 'new_vaccinations_smoothed'
    ]
    for col in vaccine_cols:
        if col in df.columns:
            df.loc[df['date'] < '2021-01-16', col] = 0
    return df
```

### Forward-Fill Missing Vaccine Data

```python
def forward_fill_vaccination_data(df):
    """
    Forward fill vaccine-related columns to handle gaps in reporting.
    """
    vaccine_cols = [
        'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated',
        'total_boosters', 'new_vaccinations', 'new_vaccinations_smoothed'
    ]
    df[vaccine_cols] = df[vaccine_cols].fillna(method='ffill')
    return df
```

### Ensure Cumulative Columns Never Decrease

```python
def enforce_cumulative_monotonicity(df):
    """
    Force cumulative counts (like cases and vaccinations) to be non-decreasing.
    """
    cumulative_cols = ['total_cases', 'total_deaths', 'total_vaccinations']
    for col in cumulative_cols:
        if col in df.columns:
            df[col] = df[col].cummax()
    return df
```

### Add Non-Vaccinated Population Column

```python
def add_non_vaccinated_column(df):
    """
    Create a new column showing estimated number of non-vaccinated people.
    """
    if 'population' in df.columns and 'people_fully_vaccinated' in df.columns:
        df['non_vaccinated'] = df['population'] - df['people_fully_vaccinated'].fillna(0)
    return df
```

### Full Preprocessing Pipeline

```python
def preprocess_covid_data(filepath):
    """
    Executes all cleaning steps in sequence and returns a cleaned DataFrame.
    """
    df = load_and_prepare_data(filepath)
    df = drop_fully_missing_columns(df)
    df = fill_stringency_index(df)
    df = reset_pre_vaccine_data(df)
    df = forward_fill_vaccination_data(df)
    df = enforce_cumulative_monotonicity(df)
    df = add_non_vaccinated_column(df)
    df.reset_index(drop=True, inplace=True)
    return df
```

### Feature Engineering

```python
def engineer_features(df):
    df = df.copy()
    df['new_cases_per_million'] = (df['new_cases'] / df['population']) * 1e6
    df['new_deaths_per_million'] = (df['new_deaths'] / df['population']) * 1e6
    df['vaccination_rate'] = (df['people_fully_vaccinated'] / df['population']) * 100
    df['non_vaccinated_ratio'] = (df['non_vaccinated'] / df['population']) * 100
    df['bed_need_estimate'] = df['new_cases'] * 0.10
    df['rolling_avg_cases'] = df['new_cases'].rolling(window=7).mean()
    df['rolling_avg_deaths'] = df['new_deaths'].rolling(window=7).mean()

    # Clean up
    df['new_cases_per_million'] = df['new_cases_per_million'].clip(lower=0)
    df['new_deaths_per_million'] = df['new_deaths_per_million'].clip(lower=0)
    df['vaccination_rate'] = df['vaccination_rate'].clip(upper=100)
    df['non_vaccinated_ratio'] = df['non_vaccinated_ratio'].clip(upper=100)
    
    return df

# Apply it
df_features = engineer_features(df)
```

```python
def plot_feature(df, column, title, ylabel, color='teal'):
    fig, ax = plt.subplots(figsize=(14, 5))
    ax.plot(df['date'], df[column], color=color)
    ax.set_title(title)
    ax.set_xlabel("Date")
    ax.set_ylabel(ylabel)
    ax.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
```

### Risk Score Modeling & Visualization

```python
# STEP 1: Risk Score Computation
def compute_risk_score(df):
    """
    Computes risk score based on:
    - new_cases_per_million (40%)
    - non_vaccinated_ratio (30%)
    - bed_need_estimate (20%)
    - case fatality ratio (10%)
    """
    df = df.copy()
    
    # Compute CFR
    df['cfr'] = (df['total_deaths'] / df['total_cases']).replace([np.inf, -np.inf], np.nan).fillna(0).clip(upper=0.1)
    
    # Normalization (Min-Max)
    def normalize(series):
        return (series - series.min()) / (series.max() - series.min())

    df['norm_cases'] = normalize(df['new_cases_per_million'])
    df['norm_non_vax'] = normalize(df['non_vaccinated_ratio'])
    df['norm_beds'] = normalize(df['bed_need_estimate'])
    df['norm_cfr'] = normalize(df['cfr'])

    # Weighted Risk Score
    df['risk_score'] = (
        0.40 * df['norm_cases'] +
        0.30 * df['norm_non_vax'] +
        0.20 * df['norm_beds'] +
        0.10 * df['norm_cfr']
    )

    # Categorize risk levels
    def categorize(score):
        if score < 0.3:
            return 'Low'
        elif score < 0.7:
            return 'Medium'
        else:
            return 'High'

    df['risk_level'] = df['risk_score'].apply(categorize)
    
    return df[['date', 'risk_score', 'risk_level']]

# STEP 2: Plot Risk Score Trend
def plot_risk_score_trend(risk_df):
    fig, ax = plt.subplots(figsize=(14, 5))
    ax.plot(risk_df['date'], risk_df['risk_score'], color='darkorange')
    ax.set_title("COVID-19 Risk Score Trend")
    ax.set_xlabel("Date")
    ax.set_ylabel("Risk Score (0 to 1)")
    ax.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# STEP 3: Plot Risk Level Categories
def plot_risk_level_classification(risk_df):
    fig, ax = plt.subplots(figsize=(14, 5))
    colors = {'Low': 'green', 'Medium': 'gold', 'High': 'red'}
    ax.scatter(risk_df['date'], risk_df['risk_score'], 
               c=risk_df['risk_level'].map(colors), alpha=0.6)
    ax.set_title("Risk Level Classification Over Time")
    ax.set_xlabel("Date")
    ax.set_ylabel("Risk Score")
    ax.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# Make sure you have already loaded and processed 'df_features'
risk_df = compute_risk_score(df_features)
plot_risk_score_trend(risk_df)
plot_risk_level_classification(risk_df)
```

## Forecasting

### ARIMA Forecasting 

```python
# Step 1: Prepare the time series
def prepare_timeseries(df):
    df_ts = df[['date', 'new_cases']].dropna().copy()
    df_ts = df_ts.set_index('date')
    df_ts = df_ts.asfreq('D')  # Ensure daily frequency
    df_ts['new_cases'] = df_ts['new_cases'].interpolate()  # Fill missing days
    return df_ts

# Step 2: Fit ARIMA and Forecast
def forecast_arima(df_ts, forecast_days=60):
    train = df_ts[:-forecast_days]
    test = df_ts[-forecast_days:]

    model = ARIMA(train, order=(5, 1, 2))  # You can tune this
    model_fit = model.fit()

    forecast = model_fit.forecast(steps=forecast_days)
    return forecast, test['new_cases']

# Step 3: Evaluate Forecast
def evaluate_forecast(actual, predicted):
    mae = mean_absolute_error(actual, predicted)
    rmse = math.sqrt(mean_squared_error(actual, predicted))
    return mae, rmse

# Step 4: Plot Forecast
def plot_forecast(actual, predicted, title="ARIMA Forecast vs Actual"):
    plt.figure(figsize=(14, 5))
    plt.plot(actual.index, actual.values, label='Actual', color='blue')
    plt.plot(actual.index, predicted.values, label='Forecast', color='orange')
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("New Cases")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# === RUN ARIMA Forecast ===
df_ts = prepare_timeseries(df_features)
forecast_arima_vals, test_vals = forecast_arima(df_ts)

# Convert to Pandas Series for plotting
forecast_series = pd.Series(forecast_arima_vals, index=test_vals.index)

# Evaluate
mae_arima, rmse_arima = evaluate_forecast(test_vals, forecast_series)

# Plot
plot_forecast(test_vals, forecast_series, "ARIMA Forecast of COVID-19 New Cases")
```

### Porphet Forecasting

```python
# 1. Prepare data for Prophet
def prepare_prophet_data(df):
    prophet_df = df[['date', 'new_cases']].dropna().copy()
    prophet_df.rename(columns={'date': 'ds', 'new_cases': 'y'}, inplace=True)
    return prophet_df

# 2. Fit Prophet and forecast
def forecast_with_prophet(df, forecast_days=60):
    model = Prophet(daily_seasonality=True)
    model.fit(df)

    future = model.make_future_dataframe(periods=forecast_days)
    forecast = model.predict(future)

    return model, forecast

# 3. Evaluate the forecast
def evaluate_prophet(actual, predicted):
    mae = mean_absolute_error(actual, predicted)
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    return mae, rmse

# 4. Custom plot with confidence intervals and legend
def plot_prophet_forecast(actual_df, forecast_df, forecast_days=60):
    plt.figure(figsize=(14, 6))
    
    # Plot actual data
    plt.plot(actual_df['ds'], actual_df['y'], label='Actual', color='blue', linewidth=2)
    
    # Forecast and confidence interval
    plt.plot(forecast_df['ds'], forecast_df['yhat'], label='Forecast', color='orange', linewidth=2)
    plt.fill_between(forecast_df['ds'], forecast_df['yhat_lower'], forecast_df['yhat_upper'],
                     color='orange', alpha=0.3, label='Confidence Interval')
    
    plt.title(" Prophet Forecast: COVID-19 Daily New Cases in India")
    plt.xlabel("Date")
    plt.ylabel("New Cases")
    plt.legend(loc='upper left')
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# === Run everything ===
# Load your dataset if not loaded yet
# df = pd.read_csv("cleaned_india_covid_data.csv")
# df['date'] = pd.to_datetime(df['date'])

# Step-by-step execution
prophet_df = prepare_prophet_data(df_features)
model, forecast_df = forecast_with_prophet(prophet_df, forecast_days=60)

# Evaluate forecast (last 60 days)
actual_vals = prophet_df['y'][-60:].reset_index(drop=True)
predicted_vals = forecast_df['yhat'][-60:].reset_index(drop=True)

mae_prophet, rmse_prophet = evaluate_prophet(actual_vals, predicted_vals)

# Plot results
plot_prophet_forecast(prophet_df, forecast_df)

# Show metrics
print(f" Prophet MAE: {mae_prophet:.2f}")
print(f" Prophet RMSE: {rmse_prophet:.2f}")
```

### LSTM Forecast vs Actual – Daily COVID-19 New Cases

```python
# 1. Prepare data for LSTM
def prepare_lstm_data(df, feature='new_cases', window_size=30, forecast_days=60):
    data = df[[feature]].dropna().values
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(data)

    X, y = [], []
    for i in range(window_size, len(scaled_data) - forecast_days):
        X.append(scaled_data[i-window_size:i])
        y.append(scaled_data[i:i+forecast_days])
    X, y = np.array(X), np.array(y)
    return X, y, scaler, data

# 2. Build and train the LSTM model
def train_lstm_model(X_train, y_train, forecast_days=60):
    model = Sequential()
    model.add(LSTM(units=64, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(Dense(units=forecast_days))
    model.compile(optimizer='adam', loss='mse')
    model.fit(X_train, y_train, epochs=20, batch_size=16, verbose=1)
    return model

# 3. Make forecast
def forecast_lstm(model, last_sequence, scaler):
    prediction_scaled = model.predict(last_sequence)
    prediction = scaler.inverse_transform(prediction_scaled.reshape(-1, 1))
    return prediction.flatten()

# 4. Evaluate forecast
def evaluate_lstm(actual, predicted):
    mae = mean_absolute_error(actual, predicted)
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    return mae, rmse

# 5. Plot the forecast
def plot_lstm_forecast(actual, predicted, dates):
    plt.figure(figsize=(14, 6))
    plt.plot(dates, actual, label='Actual', color='blue', linewidth=2)
    plt.plot(dates, predicted, label='LSTM Forecast', color='green', linestyle='--', linewidth=2)
    plt.title("LSTM Forecast vs Actual – Daily COVID-19 New Cases")
    plt.xlabel("Date")
    plt.ylabel("New Cases")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# Load your cleaned data
df = pd.read_csv("cleaned_india_covid_data.csv")
df['date'] = pd.to_datetime(df['date'])
df_lstm = df[['date', 'new_cases']].dropna().reset_index(drop=True)

# Prepare data
window_size = 30
forecast_days = 60
X, y, scaler, raw_data = prepare_lstm_data(df_lstm, feature='new_cases', window_size=window_size, forecast_days=forecast_days)

# Split into train and prediction
X_train = X[:-1]
y_train = y[:-1]
last_sequence = X[-1:]
actual_vals = raw_data[-forecast_days:].flatten()

# Train the model
lstm_model = train_lstm_model(X_train, y_train)

# Forecast
predicted_vals = forecast_lstm(lstm_model, last_sequence, scaler)

# Evaluate
mae_lstm, rmse_lstm = evaluate_lstm(actual_vals, predicted_vals)

# Plot
forecast_dates = df_lstm['date'].iloc[-forecast_days:]
plot_lstm_forecast(actual_vals, predicted_vals, forecast_dates)

# Output metrics
print(f" LSTM MAE: {mae_lstm:.2f}")
print(f" LSTM RMSE: {rmse_lstm:.2f}")
```
